```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt

import matplotlib.dates as mdates
import matplotlib.cbook as cbook

# File to Load (Remember to Change These)
file_to_load = "./Resources/County_MedianListingPrice_AllHomes.csv"
mlp_data = pd.read_csv(file_to_load, encoding='ISO-8859-1')

# Read mlp File and store into Pandas data frame
mlp_data = pd.read_csv(file_to_load, encoding='ISO-8859-1')
mlp_data.head()
```

```python
#deleting columns not required
del mlp_data['Metro'],
del mlp_data['StateCodeFIPS'],
del mlp_data['MunicipalCodeFIPS'],
del mlp_data['SizeRank']
mlp_data.head()


```

```python
# Filter by North Carolina data
mlp_nc_data = mlp_data.loc[mlp_data['State'] == 'NC'].reset_index()
mlp_nc_data.head()
```

```python
#Top counties as of the latest data - May 2019
topcounties_05_2019 = mlp_nc_data.nlargest(5, ['2019-05']) 
topcounties_05_2019
```

```python
#Axes data
year_month=list(topcounties_05_2019.iloc[:,3:].columns.values)

print (len(year_month))
print (type(year_month))

topcounties_05_2019.iloc[0:1,3:].transpose()
topcounties_05_2019.iloc[2:3,1:2]
topcounties_05_2019['RegionName'].values[0]
```

```python
#Plot the top 5 most expensive counties

plt.figure(figsize=(15,10))

plt.plot(year_month, topcounties_05_2019.iloc[0:1,3:].transpose(), label = topcounties_05_2019['RegionName'].values[0], color='blue')
plt.plot(year_month, topcounties_05_2019.iloc[1:2,3:].transpose(), label = topcounties_05_2019['RegionName'].values[1], color='purple')
plt.plot(year_month, topcounties_05_2019.iloc[2:3,3:].transpose(), label = topcounties_05_2019['RegionName'].values[2], color='orange')
plt.plot(year_month, topcounties_05_2019.iloc[3:4,3:].transpose(), label = topcounties_05_2019['RegionName'].values[3], color='red')
plt.plot(year_month, topcounties_05_2019.iloc[4:5,3:].transpose(), label = topcounties_05_2019['RegionName'].values[4], color='green')

#Plot title and legend
plt.legend(loc="best")
plt.title("Median Housing Price per County (NC State) | 2010 (Jan) - 2019 (May) | Most Expensive 5")
plt.xlabel("Months | 2010 (Jan) - 2019 (May)")
plt.ylabel("Median Housing Price ($)")

#Set Tick locations on x axis
plt.xticks(year_month[0::12], rotation=90)

#Save the image as PNG

plt.savefig("./Images/MedianHousingPricePerCountyTop5.png")
plt.show()
```

```python
#Plot the most expensive county | Chatham

plt.figure(figsize=(15,10))
plt.plot(year_month, topcounties_05_2019.iloc[0:1,3:].transpose(), label = topcounties_05_2019['RegionName'].values[0], color='blue')
plt.legend(loc="best")
plt.title("Home Sales in Chatham County")
plt.xlabel("Months | 2010 (Jan) - 2019 (May)")
plt.ylabel("Median Housing Price ($)")

#Set Tick locations on x axis
plt.xticks(year_month[0::12], rotation=90)

#Save image as PNG
plt.savefig("./Images/HomeSalesinChathamCounty.png")
plt.show()
```

```python
#Bottom 5 counties as of latest data | May 2019
bottomcounties_05_2019 = mlp_nc_data.nsmallest(5, ['2019-05']) 
bottomcounties_05_2019
```

```python
#Axes data

year_month=list(bottomcounties_05_2019.iloc[:,3:].columns.values)

print (len(year_month))
print (type(year_month))

bottomcounties_05_2019.iloc[0:1,3:].transpose()
bottomcounties_05_2019.iloc[2:3,1:2]
bottomcounties_05_2019['RegionName'].values[0]
```

```python
# Plot the least expensive (5) counties as of latest data

plt.figure(figsize=(15,10))


plt.plot(year_month, bottomcounties_05_2019.iloc[0:1,3:].transpose(), label = bottomcounties_05_2019['RegionName'].values[0], color='orange')
plt.plot(year_month, bottomcounties_05_2019.iloc[1:2,3:].transpose(), label = bottomcounties_05_2019['RegionName'].values[1], color='red')
plt.plot(year_month, bottomcounties_05_2019.iloc[2:3,3:].transpose(), label = bottomcounties_05_2019['RegionName'].values[2], color='green')
plt.plot(year_month, bottomcounties_05_2019.iloc[3:4,3:].transpose(), label = bottomcounties_05_2019['RegionName'].values[3], color='cyan')
plt.plot(year_month, bottomcounties_05_2019.iloc[4:5,3:].transpose(), label = bottomcounties_05_2019['RegionName'].values[4], color='royalblue')

#Plot title and legend
plt.legend(loc="best")
plt.title("Median Housing Price per County (NC State) | 2010 (Jan) - 2019 (May) | Least Expensive 5")
plt.xlabel("Months | 2010 (Jan) - 2019 (May)")
plt.ylabel("Median Housing Price ($)")

#Set Tick locations on x axis
plt.xticks(year_month[0::12], rotation=90)

#Save image as PNG

plt.savefig("./Images/MedianHousingPricePerCountyBottom5.png")
plt.show()
```

```python
# Plot the least expensive county | Edgecombe

plt.figure(figsize=(15,10))
plt.plot(year_month, bottomcounties_05_2019.iloc[0:1,3:].transpose(), label = bottomcounties_05_2019['RegionName'].values[0], color='orange')
plt.legend(loc="best")
plt.title("Home Sales in Edgecombe County")
plt.xlabel("Months | 2010 (Jan) - 2019 (May)")
plt.ylabel("Median Housing Price ($)")

#Set Tick Locations for xaxis
plt.xticks(year_month[0::12], rotation=90)

#Save the image as PNG
plt.savefig("./Images/HomeSalesinEdgecombeCounty.png")
plt.show()
```
